8. Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a média)

```python
#sendo a coluna 'Pressão sanguinea' do tipo 'object', criar nova coluna com os valores numéricos
#função para converter a PA em valor numérico

def separar_pressao(pressao):
  if pd.isna(pressao):
    return None
  try:
    return int(pressao.split('/')[0])
  except:
    return
```

```python
# Criando nova coluna numérica para pressão arterial
df_sono['Pressao sanguinea_num'] = df_sono['Pressão sanguinea'].apply(separar_pressao)
df_sono.head()
```

```python
# Calculando média por gênero
media_pressao_genero = df_sono.groupby('Gênero')['Pressao sanguinea_num'].mean()
media_pressao_genero
```

Considerando apenas o momento 'sistóle' que corresponde 120mmhg (quando normal), **as mulheres possuem pressão sanguínea maior que os homens**.

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
#agrupar o frequência cardiaca
alta_freq = df_sono [df_sono['Frequência cardíaca'] > 70]
baixa_freq = df_sono[df_sono['Frequência cardíaca'] <= 70]
```

```python
#calculando a média
media_alta_freq = alta_freq['Passos diários'].mean()
media_baixa_freq = baixa_freq['Passos diários'].mean()
```

```python
#exibindo
pd_pressao = pd.DataFrame({'Média Alta Frequência': [media_alta_freq], 'Média Baixa Frequência': [media_baixa_freq]})
pd_pressao.style.format('{:.2f}')
```

Pessoas com frequência cardiaca menor ou igual a 70bpm dão mais passos diários.
